```python
for i, predictor in enumerate(df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned", "Risk Weighting", "Income Band"]].columns):
                    plt.figure(i)
                    sns.set(style="whitegrid", palette="pastel")

# Define categorical columns
categorical_cols = ["BRId", "GenderId", "IAId", "Amount of Credit Cards",
                    "Nationality", "Occupation", "Fee Structure",
                    "Loyalty Classification", "Properties Owned",
                    "Risk Weighting", "Income Band"]

# Plot each categorical feature
for i, predictor in enumerate(categorical_cols):
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=predictor, order=df[predictor].value_counts().index, palette="viridis")

    plt.title(f'Distribution of {predictor}', fontsize=14, fontweight='bold')
    plt.xlabel(predictor, fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

## Numerical Analysis


```python
numerical_cols = ['Estimated Income','Superannuation Savings','Saving Accounts','Bank Loans','Bank Deposits','Foreign Currency Account','Business Lending','Credit Card Balance','Checking Accounts']

# Univariate analysis and visualisation
plt.figure(figsize=(12, 8))  # Increased size for better visibility
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i+1)
    sns.histplot(df[col],kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')

plt.tight_layout()  # Move outside the loop - prevents overlapping of subplots
plt.show()  # Single show() call after all subplots are created
```

## Heatmap


```python
numerical_cols = ['Estimated Income','Superannuation Savings','Saving Accounts','Bank Loans','Bank Deposits','Foreign Currency Account','Business Lending','Credit Card Balance','Checking Accounts']
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix,annot=True,cmap='viridis', fmt=".2f", linewidths=0.5)
```

Bank Deposits and Checking Accounts have a very strong positive correlation (0.84), suggesting users with high deposits also maintain significant balances in checking accounts.

Saving Accounts and Bank Deposits also show a strong positive correlation (0.75), indicating that customers who save more tend to deposit more as well.

Estimated Income has a moderate positive correlation with:

Superannuation Savings (0.37)

Bank Loans (0.33)

Business Lending (0.33)

Credit Card Balance (0.30)
This implies higher-income individuals are more likely to have savings and access to loans/credit.

Superannuation Savings shows low correlation with most features except Estimated Income, indicating itâ€™s more influenced by income than other financial behaviors.

Foreign Currency Account shows moderate correlations with multiple features (e.g., Bank Loans: 0.36, Bank Deposits: 0.41), suggesting international banking behavior might relate to overall financial activity.

No negative correlations are observed in the heatmap, indicating that as one feature increases, others either remain neutral or increase â€” there are no inverse relationships.

Most features have weak to moderate correlations (0.2 to 0.4), meaning they are somewhat related but not strongly dependent on one another.

Highly correlated pairs (like Bank Deposits & Checking Accounts) may introduce multicollinearity in machine learning models and could be considered for removal or dimensionality reduction (e.g., PCA).
